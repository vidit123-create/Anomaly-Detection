#### Train dataset as Analysis dataset :-

#### Test dataset as Original dataset :- 

```python
df1=pd.read_csv('C:\\Users\\arpit\\Downloads\\test_AbJTz2l.csv')
df1
```

```python
#Rename the columns:
```

```python
df1.rename({'Item_Weight':'Original_Weight','Item_Identifier':'Item_ID','Item_Visibility':'Original_Visibility','Item_MRP':'Original_MRP','Item_Type':'Original_Item_Type'},axis=1,inplace=True)
df1
```

```python
#Assign the dataframe to a new column :-
```

```python
df2=df1
df2
```

```python
#Merge both the dataframes :-
```

```python
df3=pd.merge(df,df2,on='Outlet_Identifier',how='inner')
df3
```

```python
#Now do subset on the basis Outlet_Size as Medium :-
```

```python
df4=df3[(df3['Outlet_Size_x']=='Medium')&(df3['Outlet_Size_y']=='Medium')]
df4
```

```python
#Preparation of questions :-
```

#### Item_Weight_Anomaly :-

```python
#Q. Let's find the mismatch of values according to Item_Weight and calculate the Original_Sales and Extra_Sales :-
```

```python
#Comaprision of Analysis_Weight and Original_Weight :-
```

```python
df4['Mismatched_Weight']=df4['Item_Weight']>df4['Original_Weight']
df4
```

```python
#Count :-
```

```python
df4['Mismatched_Weight'].value_counts()
```

```python
#Subset :-
```

```python
df5=df4[df4['Mismatched_Weight']==True]
df5
```

```python
# Q.Let's calculate Original_Sales for each product for a particular Outlet-OUT049 :-
```

#### Dairy Product :-

```python
#Subset :-
```

```python
df6=df5[(df5['Outlet_Identifier']=='OUT049')&(df5['Item_Type']=='Dairy')&(df5['Original_Item_Type']=='Dairy')]
df6
```

```python
#Calculate Total_Item_Sold :-
```

```python
df6['Total_Item_Sold']=df6['Item_Outlet_Sales']/df6['Item_MRP']
df6
```

```python
#Calculate Original_Sales :-
```

```python
df6['Original_Sales']=df6['Original_MRP']*df6['Total_Item_Sold']
df6
```

```python
#Q. Now let's compare both Sales column in the merged dataset.
```

```python
#Compare :-
```

```python
df6['Sales_Comparision']=df6['Item_Outlet_Sales']>df6['Original_Sales']
df6
```

```python
#Count :-
```

```python
df6['Sales_Comparision'].value_counts()
```

```python
#Subset :-
```

```python
df7=df6.loc[df6['Sales_Comparision']==True]
df7
```

```python
#Calculate Extra_Sales earned by the Shopkeeper owner  :-
```

```python
df7['Extra_Sales']=df7['Item_Outlet_Sales']-df7['Original_Sales']
df7
```

```python
#Q. Now let's calculate overall sales,total_item_sold and Extra_Sales using Groupby :-
```

```python
#Groupby :-
```

```python
df8=df7.groupby(['Outlet_Identifier','Item_ID','Mismatched_Weight']).agg({'Original_Sales':'sum','Item_Outlet_Sales':'sum','Mismatched_Weight':'count','Extra_Sales':'sum','Total_Item_Sold':'sum'})
df8
```

```python
#Data Manipulation :-
```

```python
#Rename some columns :-
```

```python
df8.rename({'Mismatched_Weight':'Weight','Total_Item_Sold':'Item_Sold'},axis=1,inplace=True)
df8
```

```python
#Assign to a new variable :-
```

```python
df9=df8
df9
```

```python
#Sort the columns :-
```

```python
df10=df9.sort_values(by='Original_Sales',ascending=False)
df10
```

```python
#Write to an excel file :-
```

```python
df10.to_excel('Item_667.xlsx')
```
