```python
#Import libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Fuel dataset:-
```

```python
df=pd.read_excel('C:\\Users\\arpit\\Downloads\\Fuel_Data_(Rev.01-25-Feb-20).xlsx')
df
```

```python
#Master dataset:-
```

```python
df1=pd.read_excel('C:\\Users\\arpit\\Downloads\\Master.xlsx')
df1
```

```python
#Merge dataset:-
```

```python
df2=pd.merge(df,df1,on='PTROCRD',how='inner')
df2
```

```python
#Clean Fuel dataset:-
```

```python
def A(data):
    data=str(data)
    data_x=data.replace('-','')
    data_y=data_x.replace(' ','')
    data_z=data_y.lower()
    return data_z
df2['fuel_clean']=df2['VEHICLE'].apply(A)
df2
```

```python
#Clean Master dataset:-
```

```python
def A(data):
    data1=str(data)
    data1_x=data1.replace('-','')
    data1_y=data1_x.replace(' ','')
    data1_z=data1_y.lower()
    return data1_z
df2['master_clean']=df2['VEHICLE NO'].apply(A)
df2
```

```python
#Comparision:-
```

```python
df2['unmatched']=df2['fuel_clean']!=df2['master_clean']
df2
```

```python
#Subsetting:-
```

```python
df3=df2[df2['unmatched']==True]
df3
```

```python
#Groupby:-
```

```python
x=df3.groupby(['TRANTYP','fuel_clean','master_clean','unmatched']).agg({'unmatched':'count','EXPAMT':'sum'})
x
```

```python
#Data Manipulation
```

```python
#Sort:-
```

```python
y=x.sort_values(by='EXPAMT',ascending=False)
y
```

```python
#Rename:-
```

```python
y.rename({'unmatched':'mismatched'},axis=1,inplace=True)
y
```

```python
#Data preparation:-
```

```python
df4=y
df4
```

```python
#Write to excel
```

```python
df4.to_excel('Fuel018.xlsx')
```
