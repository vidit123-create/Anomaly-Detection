```python
#Import libraries:-
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#fuel dataset:-
```

```python
df=pd.read_excel('C:\\Users\\arpit\\Downloads\\Fuel_Data_(Rev.01-25-Feb-20).xlsx')
df
```

```python
#Master dataset:-
```

```python
df1=pd.read_excel('C:\\Users\\arpit\\Downloads\\Master.xlsx')
df1
```

```python
#Merge dataset:-
```

```python
df2=pd.merge(df,df1,on='PTROCRD',how='inner')
df2
```

```python
#Comaparision:-
```

```python
df2['Overfilling_fuel']=df2['AU_MOTOR']>df2['ORIGINAL_TANK_CAPACITY']
df2
```

```python
#Subsetting
```

```python
df3=df2[df2['Overfilling_fuel']==True]
df3
```

```python
#Now we to calculate Extra fuel used by the employees:
```

```python
df3['Extra_fuel']=df3['AU_MOTOR']-df3['ORIGINAL_TANK_CAPACITY']
df3
```

```python
#Now we have to calculate fuel price for 1 litre 
```

```python
df3['Fuel_Price']=df3['EXPAMT']/df3['AU_MOTOR']
df3
```

```python
#Now calculate total extra fuel price
```

```python
df3['Total_extra_fuel_price']=df3['Extra_fuel']*df3['Fuel_Price']
df3
```

```python
#Now let's how many employees overfilling there tank with what amount and also how much does they cost for extra amount
```

```python
df4=df3.groupby(['PTROCRD','MODEL','Overfilling_fuel']).agg({'Overfilling_fuel':'count','Extra_fuel':'sum','Total_extra_fuel_price':'sum'})
df4
```

```python
#Rename
```

```python
df4.rename({'Overfilling_fuel':'Count_Overfilling_fuel','Extra_fuel':'Total_extra_fuel'},axis=1,inplace=True)
df4
```

```python
#Assigning to a variable:-
```

```python
df5=df4
df5
```

```python
#Sort:-
```

```python
df6=df5.sort_values(by='Total_extra_fuel',ascending=False)
df6
```

```python
#Export to a excel file
```

```python
df6.to_excel('Fuel022.xlsx')
```
