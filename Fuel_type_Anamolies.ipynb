```python
#Import libraries:-
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import Fuel dataset:-
```

```python
df=pd.read_excel('C:\\Users\\arpit\\Downloads\\Fuel_Data_(Rev.01-25-Feb-20).xlsx')
df
```

```python
#Master dataset:-
```

```python
df1=pd.read_excel('C:\\Users\\arpit\\Downloads\\Master.xlsx')
df1
```

```python
#Merge dataset:-
```

```python
df2=pd.merge(df,df1,on='PTROCRD',how='inner')
df2
```

```python
#Comparision:-
```

```python
df2['unmatched']=df2['TRANTYP']!=df2['FUEL']
df2
```

```python
#Subsetting:-
```

```python
x=df2[df2['unmatched']==True]
x
```

```python
#Groupby:-
```

```python
y=x.groupby(['PTROCRD','TRANTYP','FUEL','unmatched']).agg({'unmatched':'count','EXPAMT':'sum'})
y
```

```python
#Rename:-
```

```python
y.rename({'unmatched':'mismatched'},axis=1,inplace=True)
y
```

```python
#Assign to a variable.
```

```python
df3=y
df3
```

```python
#Sort:-
```

```python
df4=df3.sort_values(by='EXPAMT',ascending=False)
df4
```

```python
#Write to excel
```

```python
df4.to_excel('Fuel020.xlsx')
```
