#### Dairy :-

```python
#Subset :-
```

```python
df5=df4[(df4['Outlet_Identifier']=='OUT049')&(df4['Item_Type_x']=='Dairy')&(df4['Item_Type_y']=='Dairy')]
df5
```

```python
#Calculate total_Item_Sold :-
```

```python
df5['Item_Sold']=df5['Item_Outlet_Sales']/df5['Item_MRP']
df5
```

```python
#Calculate Original_Sales:-
```

```python
df5['Original_Sales']=df5['Item_Sold']*df5['Original_MRP']
df5
```

```python
#Now compare Item_Outlet_Sales and Original_Sales :-
```

```python
df5['Greater_Sales']=df5['Item_Outlet_Sales']>df5['Original_Sales']
df5
```

```python
#Now see the value_counts of Greater_Sales :-
```

```python
df5['Greater_Sales'].value_counts()
```

```python
#Subset :-
```

```python
df6=df5[df5['Greater_Sales']==True]
df6
```

```python
#Calculate Extra_Money :-
```

```python
df6['Extra_Money']=df6['Item_Outlet_Sales']-df6['Original_Sales']
df6
```

```python
#Rename the columns:-
```

```python
df6.rename({'Item_Identifier_y':'Item_ID'},axis=1,inplace=True)
df6
```

```python
#Assign to a new variable :-
```

```python
df7=df6
df7
```

```python
#Groupby :-
```

```python
df8=df7.groupby(['Outlet_Identifier','Item_ID','Mismatched_Weight']).agg({'Mismatched_Weight':'count','Item_Outlet_Sales':'sum','Original_Sales':'sum','Item_Sold':'sum','Extra_Money':'sum'})
df8
```

```python
#Rename the columns :-
```

```python
df8.rename({'Mismatched_Weight':'Weight'},axis=1,inplace=True)
df8
```

```python
#Sort the columns :-
```

```python
df9=df8.sort_values(by='Original_Sales',ascending=False)
df9
```
