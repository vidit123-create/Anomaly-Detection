```python
df8=df7[(df7['Outlet_Identifier']=='OUT049')&(df7['Item_ID']=='DRE15')]
df8
```

```python
#Groupby :-

# 1.Through this groupby we could find the various stats like Original_Sales,Item_Outlet_Sales and Various other things.

# 2.These stats we find using Outlet_Id and Item_Id and Item_Id plays a very important role in finding these stats.

# 3.Since we know that in Supermarkets the products are supplied from the various manufacturers and suppose we have a particular breakfast type and this particular breakfast type contains many breakfast products and these products contain many packets.

# 4.So here we use to combine some packets of each product type which comes under breakfast type and provide them a particular Item_ID and on the basis of that Item_ID we use to calculate all the stats using groupby.
```

```python
df8=df7.groupby(['Outlet_Identifier','Item_ID','Fat2','Fat3','Mismatched_Fat']).agg({'Item_Outlet_Sales':'sum','Original_Sales':'sum','Mismatched_Fat':'count','Extra_Money':'sum','Total_Item_Sold_Out':'sum'})
df8
```
